```python
plt.figure(figsize=(16, 4))
sns.set(style="whitegrid")
x = top_songs_2017.groupby("name")["popularity"].mean().sort_values(ascending=False).head(10)
axis = sns.barplot(x.index, x)
axis.set(title='Top Songs 2017 with Popularity',
         xlabel='Tracks', 
         ylabel='Popularity')
plt.xticks(rotation = 90)
plt.show()
```
